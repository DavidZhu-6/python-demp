```python
# Use indexing of the page number and then input number into string

def get_link_from_news_title(page_num, URL):
    for i in range(page_num):
        # for automatic looping
        current_page_num = 1 + i
        position = URL.index('page/')
        URL_with_page_num = URL[:position+5] + str(current_page_num) + URL[position+5:]

        # targeting the webpage
        URL_with_page_num = urllib.request.Request(URL_with_page_num, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'})
        source_code_form_url = urllib.request.urlopen(URL_with_page_num)
        soup = BeautifulSoup(source_code_form_url, 'html.parser', from_encoding='utf-8')

        # collecting texts and links
        for title in soup.select('h2.h5.d-inline-block.cursor-pointer'):
            url_link = title.select('a')
            text = url_link[0].get_text()
            TITLE_OF_ARTICLE.append(text)

            article_URL = url_link[0]['href'] # the reason why I add [0] between article_URL and ['href'] is each text covered with the list bracket. so we need to specify the element in the each list
            URL_OF_ARTICLE.append(article_URL)

        # giving a pause for avoiding bots
        if i % 10 == 0:
            time.sleep(2)
        if i % 10 == 0:
            print(f'{i*10}nd article complete')
```

## Final code for collecting dates and contents of each articles

```python
# Make a loop for collecting article body
# It spend much time, so wait please

def get_texts_from_news(articles):
    for i, URL in zip(range(len(articles)), articles): # the former for checking the number of iteration, the latter for inputting articles

        # targeting the webpage
        LINK = urllib.request.Request(URL, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'})
        source_code_form_url = urllib.request.urlopen(LINK)
        soup = BeautifulSoup(source_code_form_url, 'html.parser', from_encoding='utf-8')

        # to collect the paragraphs of article
        content = soup.select('main.px-4 > p')
        CONTENT = []
        for para in content:
            CONTENT.append(para.get_text())

        CONTENT = " ".join(CONTENT)
        CONTENT_OF_ARTICLE.append(CONTENT)

        # to collect the date of article
        date = soup.select('time.text-secondary.updated.me-2')
        date_of_article = date[0].get_text().strip()
        DATE_OF_ARTICLE.append(date_of_article)

        if i % 10 == 0:
            time.sleep(1)
        if i % 100 == 0:
            print(f'{i}nd article complete')
```

## How to deal with some errors

* sometimes the article was deleted or other types of articles were made (video)


* HTTPError: is no actual content
* IndexError: a function or method dose not the form or form of elements

```python
from urllib.error import HTTPError

def get_texts_from_news(articles):
    for i, URL in zip(range(len(articles)), articles): # the former for checking the number of iteration, the latter for inputting articles

        # regardless of occuring errors, keep going
        try:
            # targeting the webpage
            LINK = urllib.request.Request(URL, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'})
            source_code_form_url = urllib.request.urlopen(LINK)
            soup = BeautifulSoup(source_code_form_url, 'html.parser', from_encoding='utf-8')

            # to collect the body
            content = soup.select('div.article_right')
            content = content[0].get_text()
            content = content.replace("\\'", "'")
            content = content.replace("\n", "")
            content = content.replace("\xa0", "")
            CONTENT_OF_ARTICLE.append(content)

        # define the type of errors
        except HTTPError as e: # there is no article, so we cannot collect that article
            print(f'article {i} does not exist', e) # show me what article make an error
            CONTENT_OF_ARTICLE.append('NODATA')

        except IndexError as f: # the HTML structure of the article is different, so it cannot be collected by this crawling
            print(f'article {i} cannot be collected', f) # show me what article make an error
            CONTENT_OF_ARTICLE.append('NODATA')

        if i % 10 == 0:
            time.sleep(1)
        if i % 100 == 0:
            print(f'{i}nd article complete')
```
